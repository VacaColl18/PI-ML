```python
# Importo librerias a utilizar

import pandas as pd
import numpy as np
from ast import literal_eval
import ast
```

```python
# genero dataframe a partir del dataset

df = pd.read_csv(r'C:\Users\mpvac\OneDrive\Documents\Mis Repositorios\LABS\Proyecto 1\movies_dataset.csv')
df_credits = pd.read_csv(r"C:\Users\mpvac\OneDrive\Documents\Mis Repositorios\LABS\Proyecto 1\credits.csv")
```

```python
# genero una copia del dataframe original
df_desanidado = df.copy() 
```

# Desanidado de datos

```python
# Obtengo name de la columna genres
df_desanidado['genres'] = df_desanidado['genres'].apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
df_desanidado.head()
```

```python
# Obtengo name de la columna production_companies
df_desanidado['production_companies'] = df_desanidado['production_companies'].fillna('[]')
df_desanidado['production_companies'] = df_desanidado['production_companies'].apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
df_desanidado['production_companies'].head()
```

```python
df_desanidado.head()
```

```python
# Obtengo name de la columna belongs_to_collection
df_desanidado['belongs_to_collection'] = df_desanidado['belongs_to_collection'].fillna('[]')
df_desanidado['belongs_to_collection'] = df_desanidado['belongs_to_collection'].apply(literal_eval)
df_desanidado['belongs_to_collection'] = df_desanidado['belongs_to_collection'].apply(lambda x: x['name'] if isinstance(x, dict) else None)
```

```python
df_desanidado.info()
```

```python
# Obtengo name de la columna spoken_languages
df_desanidado['spoken_languages'] = df_desanidado['spoken_languages'].fillna('[]')
df_desanidado['spoken_languages'] = df_desanidado['spoken_languages'].apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
df_desanidado['spoken_languages'].head()
```

# Realizo cambios solicitados por el equipo de Data Science (valores nulos, cambios de formato, etc)

```python
# Reemplazo valores nulos por 0 en la columna 'budget' y 'revenue'

df_desanidado['budget'].fillna(0, inplace=True)
df_desanidado['revenue'].fillna(0, inplace=True)
```

```python
# elimino valores nulos del campo release
df_desanidado.dropna(subset=["release_date"], inplace=True)
```

```python
# Columna 'Fecha' a formato fecha
df_desanidado['release_date'] = pd.to_datetime(df_desanidado['release_date'], errors='coerce')
df_desanidado.dropna(subset=["release_date"], inplace=True)

# creo columna release_year con el año de la fecha de lanzamiento y lo paso a int
df_desanidado['release_year'] = df_desanidado['release_date'].dt.year
df_desanidado['release_year'] = df_desanidado['release_year'].astype(int)
```

```python
# elimino release_date
df_desanidado.dropna(subset=["release_date"], inplace=True)
```

```python
# paso los datos de la columna "revenue" a int
df_desanidado['revenue'] = df_desanidado['revenue'].astype(int)

# si el valor de "revenue" es nulo reemplazar por 0
df_desanidado['revenue'].fillna(0, inplace=True)
```

# Crear la columna con el retorno de inversión

```python
# Creo columna "return" con el retorno de inversion

df_desanidado['budget'] = pd.to_numeric(df_desanidado['budget'], errors='coerce')
df_desanidado['revenue'] = pd.to_numeric(df_desanidado['revenue'], errors='coerce')
df_desanidado['return'] = df_desanidado.apply(lambda row: row['revenue'] / row['budget'] if row['budget'] != 0 else 0, axis=1)
```

# Elimino columnas innecesarias

```python
# Elimino columnas innecesarias
df_desanidado.drop(['video', 'imdb_id', 'adult', 'original_title', 'poster_path', 'homepage'], axis=1, inplace=True)
```

```python
# obtengo el nombre del director

def get_director_name(crew):
    crew_list = ast.literal_eval(crew)
    for member in crew_list:
        if member['job'] == 'Director':
            return member['name']
    return None

df_credits['director'] = df_credits['crew'].apply(get_director_name)

df_credits['director']
```

```python
# Cambio los valores de la columna 'id' a tipo numérico en df_creditos y df_desanidado
df_desanidado['id'] = pd.to_numeric(df_desanidado['id'], errors='coerce')
df_credits['id'] = pd.to_numeric(df_credits['id'], errors='coerce')

# union de los dos dataframes
df_unido = pd.merge(df_desanidado, df_credits, on='id')

df_unido.info()
```
